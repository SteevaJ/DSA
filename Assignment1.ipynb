```python
import pandas as pd
data=pd.read_csv("/content/iris.csv")
data
#To read the dataset to python environment
```

```python
#To display columns in the dataset
print(data.columns)
```

```python
#To calculate the mean of each column of the dataset
mean1=data.mean()
print(mean1)
```

```python
#To check for null values present in the dataset
nvalue=data.isnull().sum()
print(nvalue)
```

```python
# Visualisation 1, relation between SL and SW ,Plot 1
import matplotlib.pyplot as plt
ax=data[data.Classification=='Iris-setosa'].plot.scatter(x='SL',y='SW',color='green',label='setosa')
data[data.Classification=='Iris-versicolor'].plot.scatter(x='SL',y='SW',color='red',label='versicolor',ax=ax)
data[data.Classification=='Iris-virginica'].plot.scatter(x='SL',y='SW',color='blue',label='virginica',ax=ax)
```

```python
#Plot 2 , to find correlation between petal length and petal width of different classifications
import matplotlib.pyplot as plt
ax=data[data.Classification=='Iris-setosa'].plot.scatter(x='PL',y='PW',color='green',label='setosa')
data[data.Classification=='Iris-versicolor'].plot.scatter(x='PL',y='PW',color='red',label='versicolor',ax=ax)
data[data.Classification=='Iris-virginica'].plot.scatter(x='PL',y='PW',color='blue',label='virginica',ax=ax)
```

```python
# hence from plot 1, we see that setosa has a moderate positive correlation between sepal length and sepalwidth and that between versicolour and virginica are hard to decipher.
# From plot 2 ,we see that setosa and versicolour has strong positive correlation and virginica has moderate positive correlation between their petal length and petal width .

#Visualisation 3 , Column plotting for finding range and frequency of column values

columns=data.columns.drop(['Classification'])
x_data=range(0,data.shape[0])
fig,ax=plt.subplots()
for column in columns:
  ax.plot(x_data,data[column],label=column)
ax.legend()
```

```python
#Visualisation 4 , Now we summarise the insights obtained about the correlation between the different values
corr=data.corr()
fig,ax=plt.subplots()
im=ax.imshow(corr.values)
import numpy as np
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)
plt.setp(ax.get_xticklabels(),rotation=90,ha='right',rotation_mode='anchor')
for i in range(len(corr.columns)):
  for j in range (len(corr.columns)):
    text=ax.text(j,i,np.around(corr.iloc[i,j],decimals=2),
                 ha="center",va="center",color="black")
```
