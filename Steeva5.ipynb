Question 4 ) since we know from the info given that the columns 'Age','Cabin' and'Embarked' have missing values in them and since only the type of 'Age' is float we are using median to fill in the missing values there.

```python
for i in ['Age']: 
  data[i]=data[i].fillna(data[i].median())
```

```python
data.isna().sum()
```

```python
data['Cabin']=data['Cabin'].fillna('Unknown')
data['Cabin'][:10]
data
```

```python
data['Embarked']=data['Embarked'].fillna(data['Embarked'].value_counts().index[0])
data
```

```python
data.isna().sum()
```

Here the missing values in cabin and embarked were fixed by introducing a new value 'unknown' in the cabin column, since cabin has a very high percentage of missing values, and that of embarked were fixed by using the most frequent value instead of the missing value.
