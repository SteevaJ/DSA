```python
import pandas as pd
import numpy as np
data=pd.read_csv("/content/StudentsPerformance.csv")
data
```

```python
#Q1)number of male and female participants
f=data['gender'].value_counts()['female']
print(f)
m=data['gender'].value_counts()['male']
print(m)
```

```python
# About students parental level of education
print(data['parental level of education'].value_counts())
```

```python
#q3 who scores most on average
data.groupby('gender').agg(mean_of_mathscore=('math score','mean'),mean_of_readingscore=('reading score','mean'),mean_of_writingscore=('writing score','mean'))
```

```python
data.groupby('test preparation course').agg(mean_of_mathscore=('math score','mean'),mean_of_readingscore=('reading score','mean'),mean_of_writingscore=('writing score','mean'))
```

```python
# We see that females on an average score more than males in reading and writing and males score more on average in math and the ones who completed the test prep course score more  on average than the ones who didn't
```
